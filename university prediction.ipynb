## Building ML model

#### first seperate the dataset

```python
## input features
x=df.iloc[:,:-1]
```

```python
x.head()
```

```python
y=df.iloc[:,-1]
```

```python
y.head(2)
```

#### now split tha dataset into training and testing dataset 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## as all the features are continues we will perform multiple linear regression

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
```

```python
x_test_scaled=scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg
```

```python
reg.fit(x_train_scaled,y_train)
```

```python
y_pred=reg.predict(x_test_scaled)
```

```python
plt.scatter(y_test,y_pred)
```

```python
linscore=(reg.score(x_test_scaled,y_test))*100
```

```python
linscore
```

#### evaluating the performance of linear reg model

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
mse=(mean_absolute_error(y_test,y_pred))
mse
```

```python
r2s=r2_score(y_test,y_pred)
r2s
```

### now to reduce the overfitting we will use Ridge regression

```python
from sklearn.linear_model import Ridge
```

```python
ref=Ridge()
```

```python
ref.fit(x_train_scaled,y_train)
```

```python
y_ridge_pred=ref.predict(x_test_scaled)
```

```python
plt.scatter(y_test,y_ridge_pred)
```

### evaluating the performance of Ridge regression

```python
mse_ridge=mean_absolute_error(y_test,y_ridge_pred)
```

```python
r2_score_ridge=r2_score(y_test,y_ridge_pred)
```

```python
print(mse)
print(r2_score_ridge)
```

### now to perform better feature selection we will use Lasso regression

##### in Lasso reg to increase our accuracy we will cross validation method to train our model

```python
from sklearn.linear_model import LassoCV
```

```python
las=LassoCV(cv=5)
```

```python
las
```

```python
las.fit(x_train_scaled,y_train)
```

```python
y_lasso_pred=las.predict(x_test_scaled)
```

```python
plt.scatter(y_test,y_lasso_pred)
```

```python
mse_lasso=mean_absolute_error(y_test,y_lasso_pred)
```

```python
r2_score_lasso=r2_score(y_test,y_lasso_pred)
```

```python
print(mse_lasso)
print(r2_score_lasso)
```

### now we will use Elastic net which is the combination of both Ridge and Lasso regression

```python
from sklearn.linear_model import ElasticNetCV
```

```python
els=ElasticNetCV(cv=10)
```

```python
els.fit(x_train_scaled,y_train)
```

```python
y_elastic_pred=els.predict(x_test_scaled)
```

```python
plt.scatter(y_test,y_elastic_pred)
```

```python
mse_els=mean_absolute_error(y_test,y_elastic_pred)
```

```python
r2_els_score=r2_score(y_test,y_elastic_pred)
```

```python
print(mse_els)
print(r2_els_score)
```

### now visualize the r2 scores of each and every model

```python
methods=["Linear reg","Ridge reg","Lasso reg","Elastic net"]
scores=np.array([r2s,r2_score_ridge,r2_score_lasso,r2_els_score])
plt.bar(methods,scores,color="#FBCEB1")
```

#### from above we can see that all models have similar accuracy
